```python
import pandas as pd
```

```python
admission_data = pd.read_csv('../data/admission/Admission_Predict_Ver1.1.csv')
```

```python
admission_data
```

```python
admission_data.loc[:, 'SOP'] = (admission_data.loc[:, 'SOP']+0.1).round()
```

```python
admission_data.loc[:, 'LOR'] = (admission_data.loc[:, 'LOR ']+0.1).round()
```

```python
admission_data.loc[:, 'Research'] = admission_data.loc[:, 'Research'].astype(int)
```

```python
admission_data.loc[:, 'Admit'] = (admission_data.loc[:, 'Chance of Admit '] > 0.65).astype(int)
admission_data.loc[:, 'Admit'].sum()
```

```python
admission_data.to_csv('../data/admission/admission.csv')
```
